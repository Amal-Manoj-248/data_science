# Aggregate function

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Flavors.csv')
df
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby('Base Flavor')[numeric_cols].mean()
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','min','count','sum'],'Texture Rating':['mean','max','min','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','min','count','sum']})
```

```python
df.groupby('Base Flavor').describe()
```

```python
df.describe()
```

# Merge,Join and Concat

```python
df1=pd.read_csv('/content/LOTR.csv')
df1
```

```python
df2=pd.read_csv('/content/LOTR 2.csv')
df2
```

```python
df1.merge(df2,how='inner',on =['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how='inner')
```

```python
df1.merge(df2,how='outer')
```

```python
df1.merge(df2,how='left')
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='cross')
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2],join='outer',axis=1)
```

```python
pd.concat([df1, df2],join='outer',axis=0)
```

```python
df1.join(df2, on='FellowshipID',how='outer', lsuffix='_caller', rsuffix='_other')
```

```python
df3=df1.set_index('FellowshipID')
df3
```

```python
df4=df2.set_index('FellowshipID')
df4
```

```python
df5=df3.join(df4, how='outer',lsuffix='_left', rsuffix='_right')
df5
```

```python
df5=df3.join(df4, how='inner',lsuffix='_left', rsuffix='_right')
df5
```
