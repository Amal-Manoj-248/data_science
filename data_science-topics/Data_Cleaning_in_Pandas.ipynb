```python
import pandas as pd
```

```python
df=pd.read_excel('/content/Customer Call List.xlsx')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns='Not_Useful_Column')
df
```

```python
df['Last_Name']=df['Last_Name'].str.strip('/123. ')
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]',' ',regex=True)
df
```

```python
df["Phone_Number"]= df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(' ','-',regex=True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan',' ')
df["Phone_Number"] = df["Phone_Number"].str.replace('N-a',' ')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
```

```python
df[['Do_Not_Contact','Paying Customer']] = df[['Do_Not_Contact','Paying Customer']].apply(lambda x: x.str.replace('Yes', 'Y') if x.dtype == 'object' else x)
df[['Do_Not_Contact','Paying Customer']] = df[['Do_Not_Contact','Paying Customer']].apply(lambda x: x.str.replace('No', 'N') if x.dtype == 'object' else x)
df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df=df.fillna(' ')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
df = df.reset_index(drop=True)
df
```
