# Task 3: Heatmap Creation

**Objective:** Visualize correlations or aggregated data using heatmaps.

**•	Dataset:** Use the flights dataset from seaborn.

**•	Task:** Create a heatmap to show the number of passengers per month and year.

o	Styling: Change the color palette to YlGnBu and add annotations to show the number of passengers in each cell.

o	Frame Customization: Customize the heatmap by drawing a frame around it with thicker lines and dotted/dashed styles.


```python
flights=sns.load_dataset("flights")
print(flights.head())
```

```python
flights=sns.load_dataset("flights")
flights= flights.pivot(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))

sns.heatmap(flights, annot=True,fmt="d",cmap="YlGnBu",
            linewidths=4,linecolor="r",linestyle="--")

plt.title("Number of Passengers per Month and Year")
```

# Task 4: Violin Plot with Customized Split and Inner Style

**Objective:** Understand different inner representations in violin plots.

**•	Dataset:** Use the tips dataset.

**•	Task:** Create a violin plot to visualize the distribution of tips across different times (lunch and dinner) and further split by gender.

o	Styling: Use inner="quartile" to visualize quartiles inside the violin plot.( adds lines inside the violin to show the quartiles of the distribution.)

o	Variation: Try using inner="stick" and observe the difference.( places lines (sticks) inside the violin plot that represent individual data points.)


```python
tips=sns.load_dataset("tips")
print(tips.head())
```

```python
tips=sns.load_dataset("tips")
plt.figure(figsize=(14, 6))
plt.suptitle("Distribution of tips across different times")
plt.subplot(1, 2, 1)
sns.violinplot(x="time", y="tip", hue="sex", data=tips, split=True, inner="quartile", palette="Set1")
plt.title("Inner: Quartile")
plt.xlabel("Time of Day")
plt.ylabel("Tip Amount")


plt.subplot(1, 2, 2)
sns.violinplot(x="time", y="tip", hue="sex", data=tips, split=True, inner="stick", palette="Set1")
plt.title("Inner: Stick")
plt.xlabel("Time of Day")
plt.ylabel("Tip Amount")

plt.show()
```

# Task 5: Combining Pair Plot with Heatmap

**Objective:** Deepen understanding by combining visualizations.

**•	Dataset:** Use the iris dataset.

**•	Task:** First, create a pair plot for the iris dataset. Then calculate the correlation matrix for the dataset and visualize it using a heatmap.

o	Styling: Use a diverging color palette for the heatmap, such as RdBu.

o	Challenge: Annotate the heatmap with the correlation values and adjust the cmap center around 0 to highlight positive and negative correlations differently.


```python
iris=sns.load_dataset("iris")
print(iris.head())
```

```python
iris=sns.load_dataset("iris")
sns.pairplot(iris,hue='species')
plt.show()

print(iris.corr(numeric_only=True))

plt.figure(figsize=(10,6))
sns.heatmap(iris.corr(numeric_only=True),annot=True,cmap='RdBu',center=0)
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```
