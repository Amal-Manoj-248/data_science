#Line plot
#Task 1: Basic Line Plot

o	Create a line graph that shows the relationship between x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].

o	Label the axes as "X-axis" and "Y-axis".

o	Set the title to "Basic Line Plot".


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")

plt.plot(x,y)
plt.show()
```

#Task 2: Line Graph with Markers

o	Create a line graph for x = [0, 1, 2, 3, 4] and y = [0, 1, 4, 9, 16].

o	Add circular markers to the points and connect them with a line.

o	Set the title as "Line Plot with Markers".


```python
x=np.array([0,1,2,3,4])
y=np.array([0,1,4,9,16])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Markers")

plt.plot(x,y,marker='o',color='r')
plt.show()
```

# Task 3: Multiple Line Plots

o	Plot two lines on the same graph:

  	Line 1: x = [1, 2, 3, 4] and y1 = [1, 4, 9, 16]

  	Line 2: x = [1, 2, 3, 4] and y2 = [2, 3, 6, 8]

o	Use different colors and line styles for each line (e.g., one dashed line and one solid line).

o	Add a legend to differentiate the lines.


```python
x=np.array([1,2,3,4])
y1=np.array([1,4,9,16])
y2=np.array([2,3,6,8])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plots")

plt.plot(x,y1,color='y',linestyle='--',label='Line 1')
plt.plot(x,y2,color='g',linestyle='-',label='Line 2')
plt.legend()
plt.show()
```

#Task 4: Subplots with Line Graphs

o	Create two separate line graphs in one figure:

	First graph: x = [1, 2, 3, 4, 5] and y = [1, 4, 9, 16, 25]

	Second graph: x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10]

o	Use plt.figure() and plt.subplot() to place the graphs side by side.


```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 4, 9, 16, 25])
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([2, 4, 6, 8, 10])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x1, y1,color='m')
plt.title('First Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(1, 2, 2)
plt.plot(x2, y2,color='y')
plt.title('Second Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

# Bar Graph Tasks:
# Task 5: Basic Bar Plot


o	Create a bar plot with categories ['A', 'B', 'C', 'D'] and values [5, 7, 3, 8].

o	Label the x-axis as "Categories" and the y-axis as "Values".

o	Add a title: "Basic Bar Plot".


```python
x=np.array(['A','B','C','D'])
y=np.array([5,7,3,8])

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Basic Bar Plot")

plt.bar(x,y,color='c')
plt.show()
```

# Task 6: Horizontal Bar Plot

o	Create a horizontal bar plot for categories ['E', 'F', 'G', 'H'] and values [8, 4, 6, 3].

o	Set the title to "Horizontal Bar Plot".


```python
x=np.array(['E','F','G','H'])
y=np.array([8,4,6,3])

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Horizontal  Bar Plot")

plt.barh(x,y,color='hotpink')
plt.show()
```

# Task 7: Bar Plot with Annotations

o	Create a bar plot with categories = ['X', 'Y', 'Z'] and values = [7, 2, 5].

o	Annotate each bar with its height.


```python
x=np.array(['X','Y','Z'])
y=np.array([7,2,5])

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot with Annotations")

for i, v in enumerate(y):
    plt.text(i, v + 0.2, str(v), ha='center', va='bottom')

plt.bar(x,y,color='skyblue')
plt.show()
```

# Task 8: Create Two Figures with Different Plots


o	In one figure, create a line plot for x = [1, 2, 3, 4] and y = [1, 4, 9, 16].

o	In a second figure, create a bar plot for labels = ['A', 'B', 'C'] and values = [5, 7, 9].


```python
x1 = np.array([1, 2, 3, 4])
y1 = np.array([1, 4, 9, 16])
x2 = np.array(['A','B','C'])
y2 = np.array([5,7,9])

plt.figure()
plt.plot(x1, y1,color='m')
plt.title('Line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.figure()
plt.bar(x2, y2,color='m')
plt.title('Bar plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

# Task 10: Save a Figure to File

o	Create a simple bar plot with labels = ['Red', 'Blue', 'Green'] and values = [10, 7, 5].

o	Save the plot to a file named "bar_plot.png" using plt.savefig().


```python
x=np.array(['Red','Blue','Green'])
y=np.array([10,7,5])

plt.bar(x,y,color=['r','b','g'])
plt.title('Bar plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.savefig('bar_plot.png')
plt.show()
```
